### Linear regression or OLS(ordinary least squares) is the simplest and most classic linear method for regression.Regression finds the parameters w and b that minimizes the mean squared error between predictions  and true regression targets(y) on training set. The mean square error is the sum of the squared differences between prediction and true values. Linear regression has no parameters which is a benefit but it also has no way to control the model complexity


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score


customer=pd.read_csv("Ecommerce Customers.csv")
X=customer[['Length of Membership']]
y=customer['Yearly Amount Spent']

#train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
lm=LinearRegression().fit(X_train,y_train)

predictions=lm.predict(X_test)

#scatter plot for test data
plt.scatter(X_test,y_test,color='blue')

#plt.plot(X_train,lm.predict(X_train),color='green')

plt.plot(X_test,lm.predict(X_test),color='red')

plt.title("Ecommerce customer(Testing set)")
plt.xlabel("Length of Membership")
plt.ylabel("Yearly Amount Spent")
plt.show()

MLR_CoDT=lm.score(X_train,y_train)
print(MLR_CoDT)

MLR_coD=lm.score(X_test,y_test)
print(MLR_coD)

score=r2_score(y_test,predictions)
print("The r2 score is",score)
```
